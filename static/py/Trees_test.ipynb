```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("../data/bulk_data.csv"))
df.head()
```

```python
df['Change In Snow Depth (in)'].describe()
```

```python
bins = [-340, 0, 341]
labels = ['Melt','None','Snow']
df['binned'] = pd.cut(df['Change In Snow Depth (in)'], bins=bins, labels=labels)
df
    
```

```python
df['Change In Snow Depth (in)'].unique()
```

```python
target = df['binned']
target_names = ['Melt','None','Snow']
```

```python
data = df.drop('binned', axis=1)
feature_names = data.columns
data.head()
```

```python
dates=data['Date'].astype('datetime64[ns]')
dates=dates.dt.dayofyear
df['Date']=dates
one_hot = pd.get_dummies(data['Station Name'])
# Drop column as it is now encoded
df1 = data.drop(['Station Name'],axis = 1)
# Join the encoded df
testing_df = df1.join(one_hot)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
