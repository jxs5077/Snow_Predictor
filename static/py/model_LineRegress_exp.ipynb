```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

# y = clean_df.loc[:20000, 'Change In Snow Depth (in)'].values.reshape(-1, 1)

y = data_df['Snow Depth (in) Start of Day Values'].values.reshape(-1, 1)
# test_df['Date']=test_df['Date'].astype('int')
# pd.get_dummies(#.values.reshape(-1, 1)
X=data_df[features_list].astype('int').values.reshape(-1, len(features_list))
# X.head()

print("Shape: ", X.shape, y.shape)
X
```

```python
# Create the model and fit the model to the data

from sklearn.linear_model import LinearRegression

### BEGIN SOLUTION

model = LinearRegression()

### END SOLUTION
```

```python
# Fit the model to the data. 
# Note: This is the training step where you fit the line to the data.
model.fit(X, y)
```

```python
# Print the coefficient and the intercept for the model

print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```
