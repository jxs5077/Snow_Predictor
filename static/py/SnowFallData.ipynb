```python
#import dependencies
import pandas as pd
import numpy as np
import glob
import os
```

```python
#bring in file path
path = r'../Data/exp_data'
#use glob to read through all files
all_files=glob.glob(path + "/*.csv")
#create empty list to fill with csv data
li =[]
#create for loop to reach through each file and append to li
for filename in all_files:
    snowDataDf=pd.read_csv(filename, index_col=None, header=0)
    li.append(snowDataDf)

snowframe = pd.concat(li, axis=0, ignore_index=True)
snowframe
```

```python
cleanSnowDf=snowframe.drop(['Unnamed: 19'],axis=1)
cleanSnowDf.head()
```
