```python
df = pd.read_csv(os.path.join("../data/bulk_data.csv"))
# df=df.replace([np.inf, -np.inf], np.nan, inplace=True)
df=df.fillna(0)
df.head()
```

```python
bins = [-341, -60, -12, -6, -3, 0,  3, 6, 12, 24, 60, 342]
labels = ['melting 28ft-5ft', 'melting 5ft-1ft','melting 12in-6in', 'melting 6in-3in',
         'melting 3in-0in', 'snowing 0in-3in', 'snowing 3in-6in', 'snowing 6in-12in', 
         'snowing 12in-24in', 'snowing 2ft-5ft','snowing 5ft-28ft']
df['binned'] = pd.cut(df['Change In Snow Depth (in)'], bins=bins, labels=labels)#.astype('O')
df
    
```

```python
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df.head()
```

```python
dates=df['Date'].astype('datetime64[ns]')
dates=dates.dt.dayofyear
df['Date']=dates
# one_hot = pd.get_dummies(df['Station Name'])
one_hot2 = pd.get_dummies(df['binned'])
# Drop column as it is now encoded
df1 = df.drop(['Station Name','County Name'],axis = 1)
df1 = df1.drop(['binned'],axis = 1)
# Join the encoded df
# testing_df = df1.join(one_hot)
testing_df = df1.join(one_hot2)
```

```python
data = testing_df.drop(['Change In Snow Depth (in)','melting 28ft-5ft','melting 5ft-1ft','melting 12in-6in', 'melting 6in-3in',
         'melting 3in-0in', 'snowing 0in-3in', 'snowing 3in-6in', 'snowing 6in-12in', 'snowing 12in-24in', 'snowing 2ft-5ft','snowing 5ft-28ft'], axis=1)
feature_names = data.columns
data.head()
```
