```python
# csv_path = "../data/2020-01-01-2021-05-23-expanded.csv"
csv_path = "../data/SnowData.csv"
df=pd.read_csv(csv_path, index_col=0)
df.head()
```

```python
df.filter(['Date', 'Station Name'])
```

```python
small_df=df.loc[df['Date']>='2020-10-01',:].reset_index(drop=True)
display(small_df)
```

```python
station_names=['Beaver Ck Village', 
'Copper Mountain', 
'Berthoud Summit', 
'Lake Eldora', 
'Loveland Basin',
'Vail Mountain',
'Glen Cove', 
'Mesa Lakes', 
'Butte', 
'Bison Lake']

short_df=small_df.loc[small_df['Station Name'].isin(station_names), :]
```

```python
short_df['Station Name'].unique()
```

```python
short_df.head()
```

```python
new_columns=[
    'date', 'station_name', 'snow_water_equiv_start_of_day', 'change_in_snow_water_equiv', 'snow_depth_start_of_day',
    'change_in_snow_depth', 'elevation', 'air_temp_avg', 'lat', 'lon','county_name'
]

short_df.columns=new_columns
short_df.head()
```

```python
station_df=station_df.reset_index()
```
